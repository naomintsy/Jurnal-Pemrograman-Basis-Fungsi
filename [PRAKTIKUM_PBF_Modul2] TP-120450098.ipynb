**1. Buatlah sebuah fungsi bernama ulangi_NIM, ulangi memiliki input sebuah bilangan skalar
a, dan mengeluarkan vektor 1xn dengan seluruh elemen nya adalah a !**

```python
a = int(input("Masukkan jumlah bilangan skalar a : "))
L = list(map(int,input("\nMasukkan bilangan skalar a : ").strip().split()))[:a]

def ulangi_098(n):
    return n * 1

print(list(map(ulangi_098, L)))
```

**2. Buatlah deret bilangan sebagai berikut dengan input n sebagai panjang deret:
1/2, −1/4,1/8, .... , (−1)^n 1/2^n**

```python
n = int(input('Jumlah panjang deret : '))
a = list(range(1,n+1))

def deret(n):
    return ((-1)**(n+1)) * (1/(2**n+1))
print(list(map(deret,a)))
```

**3. Jumlahkan deret bilangan tersebut!**

```python
from functools import reduce as r

L = list(map(deret,a)) 

def add(a,b):
    res = a+b
    print('a:',a,', b:',b, '-> a+b:',res)
    return res
print(r( add, L ))
```
