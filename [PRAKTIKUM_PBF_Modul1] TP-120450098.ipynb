Soal
Seorang mahasiswa telah menyewa beberapa buku dari perpustakaan kota. Maksimal jumlah hari
sewa buku adalah 26 hari. Buku yang telah disewa adalah sebagai berikut:
1. PBF, 100 halaman (Rp 1.000 per hari)
2. ALSTRAT, 250 halaman (Rp 1.000 per hari)
3. SWARM, 200 halaman (Rp 1.500 per hari)
4. BASDAT, 350 halaman (Rp 1.500 per hari)
5. SSD, 400 halaman (Rp 2.000 per hari)

Buku tersebut akan difotokopi oleh mahasiswa sebagai referensi bahan belajar. Mesin fotokopi
hanya mampu memfotokopi sebanyak 50 halaman dalam 1 hari. Setelah semua halaman buku
selesai difotokopi, buku-buku tersebut akan dikembalikan ke perpustakaan. Hitunglah total biaya
sewa yang harus dibayarkan oleh mahasiswa tersebut dan perkirakan berapa hari yang dibutuhkan
untuk pengembaliannya.
Buatlah program dengan memodularisasikan algoritma ke dalam fungsi-fungsi untuk permasalahan
tersebut.

```python
buku_pinjaman = [
    {"Subject": "PBF", "halaman": 100, "Harga_perhari": 1000},
    {"Subject": "ALSTRAT", "halaman": 250, "Harga_perhari": 1000},
    {"Subject": "SWARM", "halaman": 200, "Harga_perhari": 1500},
    {"Subject": "BASDAT", "halaman": 350, "Harga_perhari": 1500},
    {"Subject": "SSD", "halaman": 400, "Harga_perhari": 2000},
]
```

```python
def jumlah_halaman(buku_pinjaman):
    jum_halaman = 0
    for i in buku_pinjaman:
        hal = i['halaman']
        jum_halaman = jum_halaman + hal
    return jum_halaman
```

```python
def jumlah_hari(buku_pinjaman):
    fotokopi_perhari = 50
    jum_hari = 0
    n = jumlah_halaman(buku_pinjaman) / fotokopi_perhari
    if n <= 26:
        jum_hari = jum_hari + n
    else :
        print("Melewati batas peminjaman buku, harap kurangi salah satu jenis buku")
    return round(jum_hari)
```

```python
def total_biaya(buku_pinjaman):
    total = 0
    for biaya in buku_pinjaman:
        harga_perhari = biaya['Harga_perhari'] * jumlah_hari(buku_pinjaman)
        total += harga_perhari
    return round(total)
```

```python
print("Total biaya sewa yang harus dibayarkan : Rp.",total_biaya(buku_pinjaman))
print("Jumlah hari yang dibutuhkan untuk pengembalian semua buku sekaligus: ",jumlah_hari(buku_pinjaman))
```

```python
buku_sewa = [
    ['PBF',100,1000],
    ['ALSTRAT',250,1000],
    ['SWARM',200,1500],
    ['BASDAT',350,1500],
    ['SSD',400,2000]
]
```

```python
biaya = list(map(lambda x: x[1] / 50 * x[2], buku_sewa))
hari = list(map(lambda x: x[1]/50, buku_sewa))
```
