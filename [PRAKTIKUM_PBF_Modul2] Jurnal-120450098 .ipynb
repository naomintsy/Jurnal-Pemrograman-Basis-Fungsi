penjelasan : pertama open data, jika sequence ACT artinya kita harus melihat dan observasi 2 karakter setelahnya

```python
dat = open("DNA.txt").read() #membuka file dengan nama file yang berisikan data DNA
dat = dat[:-1] #memotong index jadi hanya 1
seq = 'ACT'
dat #menampilakan data dari 'dat'
```

```python
def append_n(dat,i,n):
    return reduce ( lambda a,b:a+b , dat[i:i+n] )
def remap(dat,seq):
    return map( lambda x: append_n(dat,x,len(seq)) , range(len(dat) - len(seq )+ 1 ) )
def count_mer(dat,seq) : #mengembalikan jumlah total elemen yang diberikan dalam sebuah string. Penghitungan dimulai dari awal string sampai akhir dan membuat funngsi lambda
    return reduce (lambda a,b: a+ (1 if b==seq else 0) , remap(dat,seq) , 0 ) 
```

```python
list(remap(dat, 'ACT'))[-1] #mengubah dna 'ACT' ke dna lain
```

```python
list(remap(dat, 'ACT')) #mengubah suatu objek ke bentuk list dan mengganti dna 1 ke dna lain
len(dat)#Fungsi len() digunakan untuk mengetahui panjang (jumlah item atau anggota) dari objek dat
```

```python
len(seq) #Fungsi len() digunakan untuk mengetahui panjang (jumlah item atau anggota) dari objek seperti sequence (string, list, tuple, range)
```

```python
sequences = ['A', 'AT' , 'GGT' , 'AAGC' , 'AGCTA']

def count_all(dat,sequences):
    return map( lambda x: count_mer(dat,x) , sequences )
res = count_all(dat,sequences) #Jumlah pengembalian berapa kali res terjadi dalam dat dan sequence
print(* res)
```

penjelasan : digunakan reverse komplemen untuk mendapatkan komplemen terbalik dan komplemen REVERSE dari 'dat' digunakan untuk mendapatkan iterator terbalik dari sequence yang diberikan

```python
def komplemen (x):
    return {'A':'T', 'T':'A' , 'C':'G' , 'G':'C'}.get(x)

def reverse_komplemen(dat):
    return map( lambda x:komplemen(x), dat)
```

```python
res = reverse_komplemen(dat)
print(*res)
```

penjelasan : WTI digunakan untuk men transpose dengan inputan W dan return ke I lalu mapping W menjadi satu dimensi. 
WT untuk melakukan agar bisa terjadi perkalian x1 dan x2 dilakukan mapping untuk setiap elemen nya.
bentuk WM harus sama dengan bentuk W

```python
import math

def aktivasi (x):
    return 1/ (1+math.exp(-x))

def WTi(W,i):
    return list(map( lambda w:w[i] , W))

def WT(W):
    return list( map( lambda i : WTi(W,i) , range(len(W[0]))) )

def XW(X,W):
    return map( lambda w: reduce( lambda a,b:a+b ,map (lambda xx,ww: xx * ww , X,w),0) ,  WT(W) ) 

def input_to_hidden(X,W):
    return list( map( lambda x:aktivasi(x) , XW(X,W) ))

def feed_forward(X,W,M):
    return input_to_hidden(input_to_hidden(X,W),M)
```

```python
X = [9 , 10 , -4]
W = [ [0.5 , 0.4] , [0.3 , 0.7] , [0.25 , 0.9] ]
M = [ [0.34] , [0.45] ]
feed_forward(X,W,M)
```
