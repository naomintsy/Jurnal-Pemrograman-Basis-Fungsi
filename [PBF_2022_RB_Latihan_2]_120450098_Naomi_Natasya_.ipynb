```python
#NO2
def max(number):
  max_value = number[0]

  for value in number :
    if value > max_value :
      max_value = value

  return max_value

a = [2,-3,5,4]
print("max number using method user definedd function is :", max(a))
print("max number using meyhod built-in function is", max (a))
```

9. Modularisasikan fungsi tersebut ( breakdown menjadi beberapa fungsi )

```python
def total (number):
  x = 0 
  for value in range(0, len(number)):
    x - x + number[value]
  return x

def avg(number):
  return total(number)/len(number)

a = [1,2,3,4]
print(avg(a))
```
