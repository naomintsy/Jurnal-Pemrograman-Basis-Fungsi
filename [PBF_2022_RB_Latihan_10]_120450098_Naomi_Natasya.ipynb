MAP VIA REDUCE

```python
L = range(10)
r(lambda a,b:a +[b], L, [])
```

```python
def mymap (f,L):
  return r(lambda a,b:a + [f(b)], L, [])

mymap(lambda x:x+10,L)
```

```python
L = range(10)
r(lambda a,b:a +[b] if b%2==0 else a, L, [])
```

3. Hitung euclidian distance dari dua vektor berikut menggunakan higher order
function!
X = [2,5,6,7,10]
Y = [-2,9,2,-1,10]

```python

X = [2,5,6,7,10]
Y = [-2,9,2,-1,10]
euclid = lambda X,Y : r (lambda a,c:a+c, map (lambda x,y : ( x-y)**2, X,Y))** 0.5
euclid (X,Y)
```

5. Buatlah deret fibonacci menggunakan higher order function!

```python
#non recursive
fibo = lambda n: r ( lambda a,b : a if b[0] <=1 else a + [a[b[0]-1] + a[b[0] -2]], enumerate ( [0,1] + list (range(1,n))), [0,1]) if n>0 else [0]
```

```python
for i in range (20):
  print ('Fibonacci =', i, fibo(i))
```

```python
fibo(10)
```

```python
#recursive

fibo_rec = lambda n: 0 if n==0 else 1 if (n==1 or n==2) else fibo_rec(n-1) + fibo_rec(n-2)
deret_fibo_rec = lambda n:list (map (lambda i : fibo_rec(i), range (n+1)))
deret_fibo_rec(10)
```
